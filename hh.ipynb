```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url="https://en.wikipedia.org/wiki/List_of_highest-grossing_Hindi_films"
data  = requests.get(url).text 
soup = BeautifulSoup(data,"html.parser")



data = []
rows = soup.find_all('tr')
# Iterate over the rows and extract the cell data
for row in rows:
    cells = row.find_all(['th', 'td'])
    row_data = [cell.get_text(strip=True) for cell in cells]
    data.append(row_data)
#data[0:15]
X=pd.DataFrame(data)
X[0:16]
```

```python
Z=X.iloc[0:16,0:6]
K=Z[:1]
K
Z.columns = Z.iloc[0]
Z=Z[1:]
Z
```

```python
import re
import numpy as np
Z['Worldwide gross'] = Z['Worldwide gross'].apply(lambda x: re.sub(r'\(.*?\)', '', x))
Z['Worldwide gross'] = Z['Worldwide gross'].apply(lambda x: re.sub(r'crore', '', x))
Z['Worldwide gross'] = Z['Worldwide gross'].apply(lambda x: re.sub(r'₹', '', x))
Z['Worldwide gross'] = Z['Worldwide gross'].apply(lambda x: re.sub(r',', '', x))
Z['Worldwide gross'].dtype
Z['Worldwide gross'] = Z['Worldwide gross'].astype('float32')
Z['Worldwide gross']
```

```python
%%sql

SELECT Film, `Worldwide gross`
FROM hmov2
WHERE `Worldwide gross` >( SELECT `Worldwide gross` FROM hmov2 WHERE Film= 'PK');
```

```python
%%sql

SELECT Year, SUM(CAST(REPLACE(REPLACE(`Worldwide gross`, '₹', ''), ',', '') AS DECIMAL)) AS TotalGross
FROM hmov2
GROUP BY Year
ORDER BY TotalGross DESC;
```
